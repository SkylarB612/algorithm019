给你 n 个非负整数 a1，a2，...，an，每个数代表坐标中的一个点 (i, ai) 。在坐标内画 n 条垂直线，垂直线 i 的两个端点分别为 (i, ai) 和 (i, 0) 。找出其中的两条线，使得它们与 x 轴共同构成的容器可以容纳最多的水。

链接：https://leetcode-cn.com/problems/container-with-most-water


```python
from typing import List
```

### 方法一：

```python
class Solution(object):
    def maxArea(self, height):
        left_index = 0    # 容器左边的索引
        right_index = len(height)-1   # 容器右边的索引
        ans = 0
        while left_index != right_index:
            left_height = height[left_index]
            right_height = height[right_index]
            product = 0
            if left_height > right_height:
                product = (right_index - left_index) * right_height
                right_index -= 1
            else:
                product = (right_index - left_index) * left_height
                left_index += 1
            ans = max(ans,product)
        return ans
```

```python
Solution().maxArea([4,3,2,1,4])
```

### 方法二：

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        i, j, containMost = 0, len(height) - 1, 0

        while (i < j):
            minHeight, i, j = (height[i], i + 1, j) if height[i] < height[j] else (height[j], i, j - 1)
            containMost = max(containMost, (j - i + 1) * minHeight)

        return containMost

```

```python
Solution().maxArea([4,3,2,1,4])
```
